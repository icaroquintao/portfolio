```python
from crewai import Agent, Task, Crew
```

```python
import os
from google.colab import userdata
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')
os.environ['OPEN_AI_MODEL'] = 'gpt-4o'
os.environ['SERPER_API_KEY'] = userdata.get('SERPER_API_KEY')
```

```python
from crewai_tools import SerperDevTool, ScrapeWebsiteTool

search_tool = SerperDevTool()
scrape_tool = ScrapeWebsiteTool()
```

```python
buscador = Agent(
    role='Buscador de Conteúdo',
    goal='Busque conteúdo online o tema mais buscado do dia de hoje 08/01/2025',
    backstory='Você está trabalhando na criação de artigos sobre os assuntos mais buscados do momento. '
              'Você vai fazer uma busca sobre informações na internet, coletá-las e agrupá-las. '
              'Seu trabalho servirá de base para o Redator de Conteúdo.',
    tools=[search_tool, scrape_tool],
    verbose=True
)
```

```python
redator = Agent(
    role='Redator de Conteúdo',
    goal='Escreva um texto sobre o tema de hoje 08/01/2025',
    backstory='Você está trabalhando na redação de artigos para o tema mais buscado do momento. '
              'Você vai utilizar os dados do Buscador de Contaúdo para escrever um texto sobre o tema que ele escolheu. '
              'Dê opnião sobre o tema mas não deixa de deixar claro que é uma opnião.',
    tools=[search_tool, scrape_tool],
    verbose=True
)
```

```python
buscar = Task(
    description="""
    1. Priorize as últimas tendências, os principais atores e as notícias mais relevantes sobre os temas mais buscados de hoje 08/01/2025.\n
    2. Identifique o que está sendo tendência e o que as pessoas mais estão buscando.\n
    3. Inclua palavras-chave de SEO e dados ou fontes relevantes.
    """,
    agent=buscador,
    expected_output="Um plano de tendências sobre o que foi mais buscado no dia de hoje,temas do momento, com as palavras mais relevantes de SEO e as últimas notícias."
)


```

```python
redigir = Task(
    description="""
    1. Use os dados coletados de conteúdo para criar um texto atraente sobre o tema.\n
    2. Incorpore palavras-chave de SEO de forma natural.\n
    3. Certifique-se de que o texto esteja estruturado de forma cativante, com uma conclusão que faça o leitor refletir.\n
    """,
    agent=redator,
    expected_output="Um texto sobre o tema mais buscado do dia"
)
```

```python
equipe = Crew(
    agents=[buscador, redator],
    tasks=[buscar, redigir],
    verbose=True
)
```

```python
resultado = equipe.kickoff()
from IPython.display import Markdown
Markdown(resultado.raw)
```
